#Topic

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('telecom_data.csv')
df.head()
```

```python
sns.set_context('talk')
plt.figure(figsize = (12,10))

sns.boxplot(y = 'Total_charges', data = df, showfliers = False)

plt.show()
```

**Description of the BOXPLOT**



*   Lowest Line is the min() val
*   Highest line is the max() val
*   the box is the Q3-Q1 (75%-25%) --> Inter Quartile Range (IQR)
*   Show the range of the max and min val

#Drawback 


*   Can't find the average
*   Can't plot hue












#Creating Boxplot for numeric features

```python
plt.figure(figsize=(16,10))
num_col = []

for col in df.columns:
    if df[col].dtypes != 'O':
        num_col.append(col)

for i in range(len(num_col)):
    plt.subplot(2,3, i+1)
    sns.boxplot(y = num_col[i], data = df, showfliers = False)

#plt.suptitle('Boxplot of the Numeric columns')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(16,10))
num_col = []

for col in df.columns:
    if df[col].dtypes != 'O':
        num_col.append(col)

for i in range(len(num_col)):
    plt.subplot(2,3, i+1)
    sns.boxplot(y = num_col[i], data = df, showfliers = False, x = df['Churn'])

#plt.suptitle('Boxplot of the Numeric columns')
plt.tight_layout()
plt.show()

```

#Boxplot with hue

```python
plt.figure(figsize=(25,12))
num_col = []

for col in df.columns:
    if df[col].dtypes != 'O':
        num_col.append(col)

for i in range(len(num_col)):
    plt.subplot(2,3, i+1)
    
    #to set a hue x and y both axis are needed
    sns.boxplot(y = num_col[i], data = df, showfliers = False, x = df['Intl_Plan'], hue = 'Churn')

#plt.suptitle('Boxplot of the Numeric columns')
plt.tight_layout()
plt.show()

```

#Violin Plot

```python
plt.figure(figsize=(25,12))
num_col = []

for col in df.columns:
    if df[col].dtypes != 'O':
        num_col.append(col)

for i in range(len(num_col)):
    plt.subplot(2,3, i+1)
    
    #to set a hue x and y both axis are needed
    sns.violinplot(y = num_col[i], data = df, x = df['Intl_Plan'], hue = 'Churn', split = True)

#plt.suptitle('Boxplot of the Numeric columns')
plt.tight_layout()
plt.show()
```
