#Scatter Plot


*   milage vs price
*   engineSize vs price
*   mpg vs price

Scatter plot helps to find co-relation between variables.
**target variable @ Y axis**





```python
sns.set_context('talk')
plt.figure(figsize=(18,14))

plt.subplot(3,1,1)
sns.scatterplot(x = 'engineSize', y = 'price', data = df)

plt.subplot(3,1,2)
sns.scatterplot(x = 'mpg', y = 'price', data = df)

plt.subplot(3,1,3)
sns.scatterplot(x = 'mileage', y = 'price', data = df,)

plt.tight_layout()
plt.show()
```

```python
sns.set_context('talk')
plt.figure(figsize=(18,12))

plt.subplot(2,2,1)
sns.scatterplot(x = 'engineSize', y = 'price', data = df, hue = 'fuelType', style='fuelType')

plt.subplot(2,2,2)
sns.scatterplot(x = 'mpg', y = 'price', data = df, hue = 'fuelType')

plt.subplot(2,2,3)
sns.scatterplot(x = 'mileage', y = 'price', data = df, hue = 'fuelType', style='fuelType')

plt.tight_layout()
plt.show()
```

#Scatterplot with a REG line --> REGplot

```python
plt.figure(figsize=(18,14))
plt.subplot(2,2,1)
sns.regplot(x = 'mileage', y = 'price', data = df)


#for changing the color of the line --> lin_kws(pass dictronary), to change the dots use scatter_kws(pass dictonary)
plt.subplot(2,2,2)
sns.regplot(x = 'mileage', y = 'price', data = df, line_kws = {'color':'red'}, scatter_kws = {'edgecolor' : 'white',
                                                                                              'color':'green'})

plt.tight_layout()
plt.show()
```

#Pair Plot

```python
#corner parameter will change the duplicate graphs
sns.pairplot(df, hue='fuelType', corner = True)
```

```python
penguins = sns.load_dataset("penguins")
sns.pairplot(penguins, hue = 'species', corner = True)
```
