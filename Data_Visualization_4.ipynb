#Topic

* Idea of **subplotting**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

**Creating New Data Frame**

```python
x = np.arange(-10,11,1)
c = 1
m = 1
y = m*x+c

#creating dataframe with random values
df = pd.DataFrame({'x':x,
                   'y':y})

df.head()
```

```python
def straight_line(x, m, c):
    df = pd.DataFrame()

    df['x'] = x
    df['y'] = m*x + c 

    return df

sl_1 = straight_line(np.arange(-10,11,1), 1, 1)
#display(stLine_1.head())

sl_2 = straight_line(np.arange(-10,11,1), -1, 1)
#display(stLine_2.head())

sl_3 = straight_line(np.arange(-10,11,1), -1, -1)

sl_4 = straight_line(np.arange(-10,11,1), 1, -1)
sl_5 = straight_line(np.arange(-10,11,1), 0, 5)
sl_6 = straight_line(np.arange(-10,11,1), 0, -5)
sl_7 = straight_line(np.arange(-10,11,1), 1, 0)
sl_8 = straight_line(np.arange(-10,11,1), -1, 0)
sl_9 = straight_line(np.arange(-10,11,1), 0, 0)
#display(stLine_3.head())
```

**Creating Subplots**

```python
sns.set_context('talk') # to change the contex of the graphs. (talk, paper, poster)

def set_label(x,y):
    plt.xlabel(x)
    plt.ylabel(y)

def set_ticks(x,y):
    plt.xticks(x)
    plt.yticks(y)

def set_title(title):
    plt.title(title)

def lineplot(x,y,title):
    plt.plot(x, y, color='red', marker='.')
    plt.hlines(0,-10,10, color='black', linestyles=':')
    plt.vlines(0,-15,15, color='black', linestyles=':')
    set_label('X','Y')
    set_ticks(np.arange(-10,11,5), np.arange(-15,16,5))
    set_title(title)
    #plt.grid()


plt.figure(figsize = (16,11))

#(3,3,1) = 3 by 3 subplot metrics , 1 = position arg
#position 1 = [1,1] row-->1, column -->1
plt.subplot(3,3,1)
lineplot(sl_1['x'], sl_1['y'],'y=mx+c where m=1 c=1')

plt.subplot(3,3,2)
lineplot(sl_2['x'], sl_2['y'],'y=mx+c where m=-1 c=1')

plt.subplot(3,3,3)
lineplot(sl_3['x'], sl_3['y'],'y=mx+c where m=-1 c=-1')

plt.subplot(3,3,4)
lineplot(sl_4['x'], sl_4['y'],'y=mx+c where m=1 c=-1')

plt.subplot(3,3,5)
lineplot(sl_5['x'], sl_5['y'],'y=mx+c where m=0 c=5')

plt.subplot(3,3,6)
lineplot(sl_6['x'], sl_6['y'],'y=mx+c where m=0 c=-5')

plt.subplot(3,3,7)
lineplot(sl_7['x'], sl_7['y'],'y=mx+c where m=1 c=0')

plt.subplot(3,3,8)
lineplot(sl_8['x'], sl_8['y'],'y=mx+c where m=-1 c=0')

plt.subplot(3,3,9)
lineplot(sl_9['x'], sl_9['y'],'y=mx+c where m=0 c=0')

'''
plt.plot(sl_1['x'], sl_1['y'], color='red', marker='.')
plt.hlines(0,-10,10, color='black', linestyles=':')
plt.vlines(0,-15,15, color='black', linestyles=':')
set_label('X','Y')
set_ticks(np.arange(-10,11,5), np.arange(-15,16,5))
set_title('y=mx+c where m=1 c=1')
#plt.grid()

#positin 2  = [1,2]
plt.subplot(3,3,2)
plt.plot(sl_2['x'], sl_2['y'], color='red', marker='.')
plt.hlines(0,-10,10, color='black', linestyles=':')
plt.vlines(0,-15,15, color='black', linestyles=':')
set_label('X','Y')
set_ticks(np.arange(-10,11,5), np.arange(-15,16,5))
set_title('y=mx+c where m=-1 c=1')

#position 3 = [1,3]
plt.subplot(3,3,3)
plt.plot(sl_3['x'], sl_3['y'], color='red', marker='.')
plt.hlines(0,-10,10, color='black', linestyles=':')
plt.vlines(0,-15,15, color='black', linestyles=':')
set_label('X','Y')
set_ticks(np.arange(-10,11,5), np.arange(-15,16,5))
set_title('y=mx+c where m=-1 c=-1')
'''
#sns.set_context('talk')
plt.tight_layout() # for avoiding overlaps
plt.show()
```

**Graph with Multiple Plot**

```python
sns.set_context('talk')
plt.figure(figsize = (15,10))

plt.plot(sl_1['x'],sl_1['y'], color = 'red', label = 'm=1 c=1')
plt.plot(sl_2['x'],sl_2['y'], color = 'blue', label = 'm=-1 c=1')
plt.plot(sl_3['x'],sl_3['y'], color = 'green', label = 'm=-1 c=-1')
plt.plot(sl_4['x'],sl_4['y'], color = 'orange', label = 'm=1 c=-1')
plt.hlines(0,-15,15, color='black', linestyles = ':')
plt.vlines(0,-10,10, color='black', linestyles = ':')
plt.xticks(np.arange(-10,11,5))
plt.yticks(np.arange(-15,16,5))
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y=mx+c')
plt.tight_layout()
plt.legend()
plt.show()
```
