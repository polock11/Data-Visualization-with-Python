```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('telecom_data.csv')
df.head()
```

```python
display(df.describe())

#information for the categorical features
display(df.describe(include = ['O']))
```

```python
sns.set_context('talk')
plt.figure(figsize = (12,8))


# histogram's X axis = for features, Y axis is auto generates for count, kde = distribution line, binwidth = bar width
sns.histplot(x = 'Total_charges', data = df, binwidth = 50, hue = 'Churn', kde = True)

#rotation = to rotate the xticks vals 90 degree
plt.xticks(np.arange(df['Total_charges'].min(), df['Total_charges'].max()+51, 50), rotation = 90)

plt.show()
```

```python
plt.figure(figsize = (12,8))

sns.histplot(y = 'Total_charges', data = df, binwidth = 50, hue = 'Churn', kde = True)
plt.yticks(np.arange(df['Total_charges'].min(), df['Total_charges'].max()+51, 50))

plt.show()
```

```python
sns.set_context('talk')
plt.figure(figsize = (12,8))

sns.histplot(x = 'Total_charges', data = df, hue = 'Churn', kde = True)
#plt.xticks(np.arange(df['Total_charges'].min(), df['Total_charges'].max()+51, 50), rotation = 90)

plt.show()
```

```python
sns.set_context('talk')
plt.figure(figsize = (12,8))

sns.histplot(x = 'Total_mins',y = 'Total_charges', data = df, hue = 'Churn')
#plt.xticks(np.arange(df['Total_charges'].min(), df['Total_charges'].max()+51, 50), rotation = 90)

plt.show()
```

```python
sns.pairplot(data = df, hue = 'Churn', corner = True)
plt.show()
```

```python
sns.set_context('talk')
plt.figure(figsize = (16,14))

plt.subplot(2,2,1)
sns.histplot(x = 'Total_charges', data = df, binwidth = 20, kde = True)

plt.subplot(2,2,2)
sns.histplot(x = 'Total_charges', data = df, binwidth = 50, color = 'green')
plt.xticks(np.arange(df['Total_charges'].min(),df['Total_charges'].max()+51, 50))

plt.subplot(2,2,3)
sns.histplot(x = 'Total_charges', data = df, binwidth = 1, kde = True, cumulative=True)
plt.title('ECDF')
plt.grid()
plt.yticks(np.arange(0,3500,250))
plt.xticks(np.arange(df['Total_charges'].min(),df['Total_charges'].max()+51, 50))

plt.subplot(2,2,4)
sns.histplot(x = 'Total_charges', data = df, binwidth = 10, kde = True, cumulative=True)
plt.yticks(np.arange(0,3500,250))
plt.title('ECDF')
plt.grid()
plt.xticks(np.arange(df['Total_charges'].min(),df['Total_charges'].max()+51, 50))

plt.tight_layout()
plt.show()
```

```python
sns.set_context('talk')
plt.figure(figsize = (16,14))

plt.subplot(2,2,1)
sns.histplot(x = 'Total_charges', data = df, binwidth = 20, kde = True, element = 'poly')

plt.tight_layout()
plt.show()
```
